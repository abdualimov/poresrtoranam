```python
import pandas as pd
import pandas_profiling
from datetime import datetime, timedelta
from ast import literal_eval
import numpy as np
import seaborn as sns
import requests
import re
rest2 = pd.read_csv('kaggle_task.csv')
rest1 = pd.read_csv('main_task.csv')
rest1['sample'] = 1
rest2['sample'] = 0
rest2['Rating'] = 0
rest = rest2.append(rest1, sort = False).reset_index(drop = True)
```

```python
rest['link'] = rest['URL_TA'].apply(lambda x: 'http://tripadvisor.com' + x)
```

```python
def poisk(s):
    print(rest.loc[rest['link'] == s].index)
    try:
        return re.findall(r"row_num \">(\d+?)<", requests.get(s).text)
        
    except:
        return 0
        
```

```python
rest['WOW'] = rest['link'].apply(poisk)
```

```python
rest.to_csv('rest.csv', index=False)
```
